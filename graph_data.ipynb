```python
symbol = 'AAPL'
data = vbt.AlpacaData.download(symbol,
                               start='2021-10-10',
                               end='2021-10-15',
                               limit=1000,
                               timeframe='15m')
inspect.signature(data.get().vbt.ohlcv.plot)
data.get().vbt.ohlcv.plot(plot_type='candlestick')
```

```python
data = data.get('Close')
windows = np.arange(2, 101)
fast_ma, slow_ma = vbt.MA.run_combs(data, window=windows, r=2, short_names=['fast', 'slow'])
entries = fast_ma.ma_crossed_above(slow_ma)
exits = fast_ma.ma_crossed_below(slow_ma)

pf = vbt.Portfolio.from_signals(data, entries, exits, init_cash=1000, freq='15m')
pf.total_profit().vbt.heatmap(x_level='fast_window', y_level='slow_window', symmetric=True).show()
```
